### Step 3 — Test semantic search via the RAG tool

```python
import json
from energy_advisor.tools.rag import search_energy_tips

test_queries = [
    "best time to charge electric vehicle",
    "maximize solar self-consumption",
    "reduce electricity bill peak hours",
]

for query in test_queries:
    result = search_energy_tips.invoke({"query": query, "max_results": 2})
    print(f"\nQuery: '{query}'")
    if "error" in result:
        print(f"  ERROR: {result['error']}")
    else:
        for tip in result["tips"]:
            print(f"  [{tip['relevance_score']}] {tip['content'][:120]}...")
```
